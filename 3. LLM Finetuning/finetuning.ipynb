## Finetuning - 

Adapting a pretrained model to a specific task by training the model on additional data

## Types - 

1. Instruction Tuning

2. Classification Tuning

### Instruction FineTuning 

- Larger datasets
- Greater Computational Power

Types - Full finetuning, Parameter Efficient Finetuning like LoRA and QLoRA
