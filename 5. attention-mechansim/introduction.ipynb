#### Why attention is needed??

- The problem with modelling long seuences



- To address this issue that we cannot translate text word by word, it is common to use a neural netwrok with two submodules - 
   1. Encoder (Reads and processes text)
   2. Decoder (Translates text)

- Before transfomers, Recurrent Neural Network(RNNS) were the most popular encoder-decoder rchitecture for language translation

## RNN

Input text 
------> 
Encoder(proceses input text sequentially) 
------> 
Upadtes hidden state at each step
------> 
Final hidden state (encoder tries to capture sentence meaning)
------>
Decoder uses this final hidden state to generate translated sentence (one word at a time) - Decoder aslo updates its hidden state at each tim

- Encoder
1. input1 + hiddenstate0 -> hiddenstate1
2. input2 + hiddenstate1 -> hiddenstate2
3. input3 + hiddenstate2 -> hiddenstate3
4. input4 + hiddenstate3 -> hiddenstate4

- Decoder
1. hiddentstate4+ inputs -> generates one word at a time

### Disdavantages of RNN-

1. RNN must remember the entire encoded input in a single hidden state before passing it to the decoder

2. Decoder only has access to one final hidden state. This leads to a loss of context, especially in complex sentences where dependencies might span long distances (longer dependencies). For long sequences this increases pressure on the decoder



## Bahdanau Attention mechanism for RNNs

- Modifies the encoder decoder RNN such that the decoder can selectively access different parts of the input sequence at each decoding step





- using an attention mechanism, the text generating decoder part of the network can access all input tokens selectively. This enables that some input tokens are more important than others for generating a given output token

- Dynamic focus on different parts of input sequence allows models to learn long range dependencies more effectively

RNN(1980) --> LSTMs(1997) --> Attention(2014) --> Transformers(2017)

## Self Attention

- Self attention is a mechanism that allows each position of the input sequence to attend to all positions in the same sequnece  when computing the representation of sequence

- Key component of contemporary mordern LLMs based on the transformer architecture such as the GPT series



- Self - Refers to mechanism or ability to compute attention weights by relating different positions in a single input sequence

- Traditional attention mechanisms- The focus is on relationships between elements of 2 different sequneces
